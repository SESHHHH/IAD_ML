#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
# Путь к разархивированной папке с данными
path = '/content/Data'

# Создание DataFrame для хранения общей таблицы, таблиц с информацией, оценками
df = pd.DataFrame()
df_info = pd.DataFrame()
df_marks = pd.DataFrame()

# Получаем список файлов внутри папки
file_list = sorted(os.listdir(path))
# print(file_list)
for file in file_list:
    data = pd.read_csv(f"/content/Data/{file}")
    if file.startswith('Students_info'):
        # Объединение информационных таблиц в одну
        df_info = pd.concat([df_info, data])
    else:
        # Объединение таблиц с оценками в одну
        df_marks = pd.concat([df_marks, data])


df = pd.merge(df_info, df_marks, on='index', how='left')
df.head(10)
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark > 90:
        return 'A'
    elif 80 <= average_mark <= 90:
        return 'B'
    elif 70 <= average_mark < 80:
        return 'C'
    elif 60 <= average_mark < 70:
        return 'D'
    elif 50 <= average_mark < 60:
        return 'E'
    else:
        return 'F'
```

```python
# Создание столбца Grade и добавление в него новой характеристики average_mark путем передачи в функцию GetGrade среднего значения оценок построчно
df['Grade'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)
```
